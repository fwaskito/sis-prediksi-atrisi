### Get data from DB

```python
from db import MysqlServer
import pandas as pd

db = MysqlServer()
```

```python
conn = db.connect(db_name='sppk')
query = 'get_train_data'
cursor = conn.cursor()
cursor.callproc(query)
for result in cursor.stored_results():
    train_ls = result.fetchall()
db.close_cursor(conn, cursor)
db.close_connection(conn)

column_names = ['id', 'attrition', 'age', 'department_name',\
        'dist_from_home', 'edu', 'edu_field', 'env_sat', 'job_sat',\
        'marital_stst','num_comp_worked', 'salary', 'wlb', 'years_at_comp']
train_df = pd.DataFrame(train_ls, columns=column_names)
train_df
```

```python
conn = db.connect(db_name='sppk')
query = 'get_test_data'
cursor = conn.cursor()
cursor.callproc(query)
for result in cursor.stored_results():
    test_ls = result.fetchall()
db.close_cursor(conn, cursor)
db.close_connection(conn)

column_names = ['id', 'attrition', 'age', 'department_name',\
        'dist_from_home', 'edu', 'edu_field', 'env_sat', 'job_sat',\
        'marital_stst','num_comp_worked', 'salary', 'wlb', 'years_at_comp']
test_df = pd.DataFrame(test_ls, columns=column_names)
test_df
```

```python
db.close_cursor(conn, cursor)
db.close_connection(conn)
```
