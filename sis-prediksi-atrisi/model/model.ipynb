```python
import pandas as pd
# Iris dataset
# dataset_path = "../data/iris.csv"
# IBM attrition dataset
data_path = "data/ibm-attrition-dataset.csv"
employees_df = pd.read_csv(data_path)
employees_df
```

#### Attribut bertipe nominal

```python
object_df = employees_df.select_dtypes(include=['object']).copy()
object_df.drop('Attrition', axis=1, inplace=True)
object_df.head()
```

```python
object_df.nunique()
```

```python
print("> Department:\n", object_df["Department"]\
                            .value_counts())
print("\n> EducationField:\n", object_df["EducationField"]\
                            .value_counts())
print("\n> MartialStatus:\n", object_df["MaritalStatus"]\
                            .value_counts())
```

#### Encoding Data Kategorik (Nominal)

```python
categorical_names = object_df.columns
categorical_names
```

```python
for i in range(len(categorical_names)):
    name = categorical_names[i]
    object_df[name] = object_df[name].astype('category')
    object_df[name] = object_df[name].cat.codes

object_df.head()
```

```python
print("> Department:\n", object_df["Department"]\
                            .value_counts())
print("\n> EducationField:\n", object_df["EducationField"]\
                            .value_counts())
print("\n> MartialStatus:\n", object_df["MaritalStatus"]\
                            .value_counts())
```

#### Ganti data nominal dalam dataframe dengan hasil encoding

```python
for i in range(len(categorical_names)):
    name = categorical_names[i]
    employees_df[name] = object_df[name]

employees_df
```

```python
employees_df["Attrition"] = employees_df["Attrition"]\
                                        .astype('category')
```

```python
employees_df.info()
```

#### Sperasi Dataset

```python
from sklearn.model_selection import train_test_split

X = employees_df.drop('Attrition', axis=1)
y = employees_df.Attrition

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                            test_size=0.30,
                                            random_state=1)
print(f"> Banyak data latih: {len(X_train)}")
print(f"> Banyak data uji: {len(X_test)}")
```

#### Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(f"\n> X_train scaled:\n{X_train}")
print(f"\n> y_train scaled:\n{X_train}")
```

#### Testing

```python
from model import KnnClassifier

knn = KnnClassifier()
y_predict = knn.get_prediction(X_train, X_test,
                                y_train, y_test, k=5)
```

```python
from sklearn.metrics import accuracy_score

#print(f"\n> Prediction result:\n {y_predict}")
print(f"\n> Accuracy: {accuracy_score(y_test, y_predict)}")
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

print("\n> Confusion matrix:\n", confusion_matrix(y_test, y_predict))
print("\n> Report:\n", classification_report(y_test, y_predict))
```
